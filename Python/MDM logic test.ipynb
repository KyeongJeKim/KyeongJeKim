```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# make random nodes 
logistics_graph = {
    'Source': {'A': 2, 'B': 9},
    'A': {'Source': 2, 'C': 7},
    'B': {'Source': 3, 'C': 3, 'D': 4},
    'C': {'A': 4, 'B': 2, 'D': 3, 'E': 5},
    'D': {'B': 4, 'C': 3, 'E': 2},
    'E': {'C': 5, 'D': 2, 'Destination': 6},
    'Destination': {'E': 6}
}
```

```python
# Establish a graph feature
G = nx.Graph()
```

```python
# logic between nodes
for node, neighbors in logistics_graph.items():
    for neighbor, weight in neighbors.items():
        G.add_edge(node, neighbor, weight=weight)
```

```python
# Calculate optimize path
shortest_path = nx.shortest_path(G, source='Source', target='Destination', weight='weight')
```

```python
# Visualize
plt.figure(figsize=(12,8))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels = True, node_size = 700, node_color = 'lightgreen', font_size = 12)
edge_labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edges(G,pos, edgelist = [(shortest_path[i], shortest_path[i+1])
                                          for i in range(len(shortest_path)-1)],
                                          width=4, edge_color='red')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=12, label_pos=0.5, font_color='red', rotate= False)
print("Optimized Path:", shortest_path)
plt.title('Optimize Route Test')
plt.axis('off')
plt.show()
```
